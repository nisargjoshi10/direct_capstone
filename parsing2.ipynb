```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import csv
import glob
import codecs
```

```python
name_list = glob.glob("A_files/*.html")
```

```python
html_stuff =[]
for i in name_list:
    op=codecs.open(i,'r')
    soup = BeautifulSoup (op, 'lxml')
    html_stuff.append(soup)
```

```python
def parsing_files(tlist):
    chem = tlist.find('table')
    chem_name = chem.tbody.tr.b.text # name of the chemical 
    chem_info = chem.tbody.tr.td
    list_temp=[]

    for div in chem_info.find_all('div'): #2nd level
        temp =  div.dl.dd
        for table in temp.find_all('table'): #3rd level 
            temp_1 = table.tbody.tr
        #         list_temp = []
            for td in temp_1.find_all('td'): #4rd level -- actrual contents
                list_temp.append(td.text)
    temp_properties2 = {}

    for t_in in range(0, len(list_temp), 2):
        temp_properties2[list_temp[t_in]] = list_temp[t_in + 1]
    return temp_properties2
```

```python
prop_list=[]
for i in html_stuff:
    rty=parsing_files(i)
    prop_list.append(rty)
```

```python
output1 = pd.DataFrame()
for i in prop_list:
    output1 = output1.append(i, ignore_index=True)
```
